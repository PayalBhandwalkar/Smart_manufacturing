```python
import json
import time
import ssl
import paho.mqtt.client as mqtt
import random
```

```python
# AWS IoT Core Details
MQTT_ENDPOINT = "a1vmj48zhnr4w-ats.iot.us-east-1.amazonaws.com"  # Replace with your AWS IoT endpoint
MQTT_PORT = 8883
MQTT_TOPIC = "smart_machine/data"
```

```python
# Paths to your certificates
CERTIFICATE_PATH = "/Users/payalbhandwalkar/Desktop/Projects/device.pem.crt"
PRIVATE_KEY_PATH = "/Users/payalbhandwalkar/Desktop/Projects/private.pem.key"
ROOT_CA_PATH = "/Users/payalbhandwalkar/Desktop/Projects/AmazonRootCA1.pem"
```

```python
# MQTT Callback Functions
#def on_connect(client, userdata, flags, rc):
#print(f"Connected with result code {rc}")

def on_connect(client, userdata, flags, rc):
  if rc == 0:
    print("✅ Connected to AWS IoT")
  else:
    print(f"⚠️ Connection failed with code {rc}")
```

```python
def on_publish(client, userdata, mid):
    print(f"✅ Message published, MID: {mid}")
```

```python
# Set up MQTT Client
client = mqtt.Client()
client.tls_set(ROOT_CA_PATH, certfile=CERTIFICATE_PATH, keyfile=PRIVATE_KEY_PATH, tls_version=ssl.PROTOCOL_TLSv1_2)

client.on_connect = on_connect
client.on_publish = on_publish

client.connect(MQTT_ENDPOINT, MQTT_PORT, 60)
client.loop_start()  # Keep MQTT client running in a separate thread
```

```python
# Simulating real-time sensor data
while True:
    sensor_data = {
        "temperature": round(random.uniform(20, 100), 2),
        "humidity": round(random.uniform(30, 90), 2),
        "vibration": round(random.uniform(0.1, 1.5), 2),
        "anomaly_detected": random.choice([True, False])
    }

    payload = json.dumps(sensor_data)
    print(f"Publishing: {payload}")
    
    client.publish(MQTT_TOPIC, payload)
    
    time.sleep(5)  # Send data every 5 seconds
```
